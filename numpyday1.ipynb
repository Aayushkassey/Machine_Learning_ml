###Speed comparision of operation on python list and numpy array

```python
import time

list1 = list(range(1000000))
sum = 0

# Using python list
start_time = time.time()
for num in list1:
  sum += num
end_time = time.time()

print(f"Sum of list1: {sum}")
print(f"Time taken on python list: {end_time-start_time}")
```

```python
np_array = np.array(range(1000000))

# Using numpy array
start_time = time.time()
sum = np.sum(np_array)
end_time = time.time()

print(f"Sum of np_array: {sum}")
print(f"Time taken on numpy array: {end_time-start_time}")
```

###NumPy Arrays Dimensions with Indexing

1D array Indexing

```python
x = np.array([1, 2, 3, 4, 5])
print(x)
```

```python
print(x[0])   # first value
print(x[3])   # fourth value
print(x[-1])  # last value
```

2D Array Indexing

```python
class_marks = np.array(
          [[40, 50, 60],
          [70, 80, 90],
          [55, 65, 75]])

print(class_marks)
```

Row 0 → [40, 50, 60]<br>
Row 1 → [70, 80, 90]<br>
Row 2 → [55, 65, 75]<br>
Indexing Format: array[row_index][column_index]

```python
print(class_marks[0][0])  # Row 0, Column 0 → 40
print(class_marks[1][2])  # Row 1, Column 2 → 90
print(class_marks[2][1])  # Row 2, Column 1 → ?
```

3D indexing

```python
school_marks = np.array(
              [[[40, 50],
                [60, 70],
                [50, 60]],

              [[80, 85],
               [90, 95],
               [90, 100]]])

print(school_marks)
```

```python
school_marks.shape
```

Class 0<br>
Student 0 → [40, 50]<br>
Student 1 → [60, 70]<br>
Student 2 → [50, 60]<br>
<br>
Class 1<br>
Student 0 → [80, 85]<br>
Student 1 → [90, 95]<br>
Student 2 → [90, 100]<br>
Indexing: array[class][row][column]

```python
print(school_marks[0][0][1])  # Class 0, Student 0, Subject 1 → 50
print(school_marks[1][1][0])  # Class 1, Student 1, Subject 0 → 90
```

```python
print(school_marks[1, 2, 1])  # Class 1, Student 2, Subject 1 → 100
```

```python
# Decorators: enhance the base function
import time

def timer(base_fn):
  def enhanced_fn():
    start_time = time.time()
    base_fn()
    stop_time = time.time()
    print(f"Time taken: {stop_time - start_time} seconds")
  return enhanced_fn

@timer
def brewtea():
  print("Making Tea")
  time.sleep(1)
  print("Tea is ready")

brewtea()
```
